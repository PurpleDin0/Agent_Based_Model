# Display the plots generated by the model


```
from IPython.display import Image

for file in os.listdir(output_path):
    if file.endswith(".png"):
        display(Image(filename=file))
```

```
import pandas as pd

# Create a dictionary of pandas dataframes for all of the csv files
data_dict = {}
for file in os.listdir(output_path):
    if file.endswith(".csv"):
        data_dict[file] = pd.read_csv(file, index_col=0)

# Print the summary information from the dataframes
for df in data_dict.keys():

  txt1 = df 
  txt2 = "summary info"
  print("\n", "*"*(26-(len(txt1))), txt1, "*"*(26-(len(txt1))))
  print("\n", "* "*(26-(len(txt2))), txt2, " *"*(26-(len(txt2))))
  print(data_dict[df])
```
